```python
# ðŸ“˜ Student Grade Prediction System (GUI Version)

import tkinter as tk
from tkinter import messagebox
import matplotlib.pyplot as plt

# Function to calculate and show result
def predict_grade():
    try:
        internal = float(entry_internal.get())
        attendance = float(entry_attendance.get())
        assignment = float(entry_assignment.get())
        final_marks = float(entry_final.get())
        late = var_late.get()
        
        # Validation
        if internal > 30 or attendance > 10 or assignment > 10 or final_marks > 50:
            messagebox.showerror("Error", "Please enter marks within valid limits.")
            return
        
        # Apply penalty if assignment is late
        if late == "Yes":
            assignment = assignment - 2  # 2 mark penalty
            if assignment < 0:
                assignment = 0
        
        total = internal + attendance + assignment + final_marks
        percentage = (total / 100) * 100

        # Grade logic
        if percentage >= 90:
            grade = "A+"
        elif percentage >= 80:
            grade = "A"
        elif percentage >= 70:
            grade = "B"
        elif percentage >= 60:
            grade = "C"
        elif percentage >= 50:
            grade = "D"
        else:
            grade = "Fail"
        
        messagebox.showinfo("Prediction Result",
                            f"Total Marks: {total}/100\nPercentage: {percentage:.2f}%\nGrade: {grade}")
        
        # Plot student's progress
        categories = ['Internal (30)', 'Attendance (10)', 'Assignment (10)', 'Final (50)']
        marks = [internal, attendance, assignment, final_marks]
        
        plt.figure(figsize=(6,4))
        plt.bar(categories, marks)
        plt.title(f"Student Progress Chart ({grade})")
        plt.ylabel("Marks")
        plt.ylim(0, 50)
        plt.show()
    
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numeric values.")

# GUI setup
root = tk.Tk()
root.title("Student Grade Prediction System")
root.geometry("400x420")
root.configure(bg="#E8F0FE")

tk.Label(root, text="ðŸŽ“ Student Grade Prediction System", font=("Arial", 14, "bold"), bg="#E8F0FE").pack(pady=10)

# Input fields
tk.Label(root, text="Internal Marks (out of 30):", bg="#E8F0FE").pack()
entry_internal = tk.Entry(root)
entry_internal.pack()

tk.Label(root, text="Attendance Marks (out of 10):", bg="#E8F0FE").pack()
entry_attendance = tk.Entry(root)
entry_attendance.pack()

tk.Label(root, text="Assignment Marks (out of 10):", bg="#E8F0FE").pack()
entry_assignment = tk.Entry(root)
entry_assignment.pack()

tk.Label(root, text="Assignment Submitted Late? (Yes/No):", bg="#E8F0FE").pack()
var_late = tk.StringVar(value="No")
tk.OptionMenu(root, var_late, "Yes", "No").pack()

tk.Label(root, text="Final Exam Marks (out of 50):", bg="#E8F0FE").pack()
entry_final = tk.Entry(root)
entry_final.pack()

tk.Button(root, text="Predict Grade", command=predict_grade, bg="#4CAF50", fg="white", font=("Arial", 11, "bold")).pack(pady=15)

root.mainloop()
```
