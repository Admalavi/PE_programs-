```python
# ======================================================
#  Student Performance Prediction System (Full GUI Version)
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from tkinter import *
from tkinter import messagebox

# -------------------------------
# Global Variables
# -------------------------------
data = {"Study_Hours": [], "Attendance": [], "Assignments": [], "Final_Score": []}
model = None

# -------------------------------
# Function to Add Student Data
# -------------------------------
def add_data():
    try:
        h = float(entry_hours.get())
        a = float(entry_attendance.get())
        s = float(entry_assignments.get())
        f = float(entry_score.get())

        data["Study_Hours"].append(h)
        data["Attendance"].append(a)
        data["Assignments"].append(s)
        data["Final_Score"].append(f)

        entry_hours.delete(0, END)
        entry_attendance.delete(0, END)
        entry_assignments.delete(0, END)
        entry_score.delete(0, END)

        messagebox.showinfo("Success", "Student data added successfully!")

    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers!")

# -------------------------------
# Function to Train Model and Show Plots
# -------------------------------
def train_and_visualize():
    global model
    if len(data["Study_Hours"]) < 3:
        messagebox.showwarning("Warning", "Please add at least 3 student records before training.")
        return

    df = pd.DataFrame(data)
    X = df[["Study_Hours", "Attendance", "Assignments"]]
    y = df["Final_Score"]

    model = LinearRegression()
    model.fit(X, y)

    messagebox.showinfo("Training Complete", " Model trained successfully!")

    # --- Visualization ---
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    sns.scatterplot(x="Study_Hours", y="Final_Score", data=df)
    plt.title("Study Hours vs Final Score")

    plt.subplot(1, 2, 2)
    sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")

    plt.tight_layout()
    plt.show()

# -------------------------------
# Function to Predict New Students Score
# -------------------------------
def predict_score():
    global model
    if model is None:
        messagebox.showwarning("Warning", "Please train the model first!")
        return

    try:
        h = float(entry_hours.get())
        a = float(entry_attendance.get())
        s = float(entry_assignments.get())

        pred = model.predict([[h, a, s]])[0]

        if pred < 50:
            msg = " Work harder! Focus more on study and attendance."
        elif pred < 75:
            msg = " Good job! You can improve more."
        else:
            msg = " Excellent! Keep it up!"

        messagebox.showinfo("Prediction Result",
                            f"Predicted Final Score: {pred:.2f}\n{msg}")

    except ValueError:
        messagebox.showerror("Error", "Please enter valid numeric values!")

# -------------------------------
# GUI Design
# -------------------------------
root = Tk()
root.title(" Student Performance Prediction System")
root.geometry("420x500")
root.config(bg="#eef3ff")

title = Label(root, text="Student Performance Predictor", font=("Arial", 16, "bold"), bg="#eef3ff", fg="#1a237e")
title.pack(pady=10)

# --- Input Fields ---
Label(root, text="Study Hours:", font=("Arial", 11), bg="#eef3ff").pack()
entry_hours = Entry(root, width=30)
entry_hours.pack()

Label(root, text="Attendance (%):", font=("Arial", 11), bg="#eef3ff").pack()
entry_attendance = Entry(root, width=30)
entry_attendance.pack()

Label(root, text="Assignments Submitted:", font=("Arial", 11), bg="#eef3ff").pack()
entry_assignments = Entry(root, width=30)
entry_assignments.pack()

Label(root, text="Final Score (out of 100):", font=("Arial", 11), bg="#eef3ff").pack()
entry_score = Entry(root, width=30)
entry_score.pack()

# --- Buttons ---
Button(root, text=" Add Student Data", command=add_data, bg="#4a90e2", fg="white",
       font=("Arial", 11, "bold"), relief="ridge").pack(pady=10)

Button(root, text=" Train & Show Plots", command=train_and_visualize, bg="#2e7d32", fg="white",
       font=("Arial", 11, "bold"), relief="ridge").pack(pady=10)

Button(root, text=" Predict Score", command=predict_score, bg="#f57c00", fg="white",
       font=("Arial", 11, "bold"), relief="ridge").pack(pady=10)

Button(root, text=" Exit", command=root.destroy, bg="#c62828", fg="white",
       font=("Arial", 11, "bold"), relief="ridge").pack(pady=10)

root.mainloop()
```
