```python
# ======================================================
# ðŸ“˜ Student Performance Prediction System (Easy Version)
# ======================================================

# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import tkinter as tk
from tkinter import messagebox

# ======================================================
# Step 1: Take Input Dataset from User
# ======================================================
print("Enter number of students:")
n = int(input("Number of records: "))

study_hours = []
attendance = []
assignments = []
final_score = []

for i in range(n):
    print(f"\nEnter details for Student {i+1}:")
    study_hours.append(float(input("Study Hours: ")))
    attendance.append(float(input("Attendance (%): ")))
    assignments.append(float(input("Assignments Submitted: ")))
    final_score.append(float(input("Final Score: ")))

# Create DataFrame
data = pd.DataFrame({
    "Study_Hours": study_hours,
    "Attendance": attendance,
    "Assignments": assignments,
    "Final_Score": final_score
})

print("\nâœ… Data Collected Successfully!\n")
print(data)

# ======================================================
# Step 2: Train Model
# ======================================================
X = data[["Study_Hours", "Attendance", "Assignments"]]
y = data["Final_Score"]

model = LinearRegression()
model.fit(X, y)

print("\nâœ… Model trained successfully!")

# ======================================================
# Step 3: Create Simple GUI for Prediction
# ======================================================
def predict_score():
    try:
        h = float(entry_hours.get())
        a = float(entry_attendance.get())
        s = float(entry_assignments.get())
        prediction = model.predict([[h, a, s]])[0]

        # Basic suggestion (Expert logic)
        if prediction < 50:
            msg = "âš ï¸ Work harder! Focus on studies and attendance."
        elif prediction < 75:
            msg = "ðŸ‘ Good! You can do even better."
        else:
            msg = "ðŸŒŸ Excellent performance!"

        messagebox.showinfo("Prediction Result",
                            f"Predicted Final Score: {prediction:.2f}\n{msg}")

    except:
        messagebox.showerror("Error", "Please enter valid numbers!")

# GUI Design
root = tk.Tk()
root.title("ðŸŽ“ Student Performance Predictor")
root.geometry("350x350")
root.config(bg="#f0f8ff")

tk.Label(root, text="Enter Student Details", font=("Arial", 14, "bold"), bg="#f0f8ff").pack(pady=10)

tk.Label(root, text="Study Hours:", bg="#f0f8ff").pack()
entry_hours = tk.Entry(root, width=25)
entry_hours.pack()

tk.Label(root, text="Attendance (%):", bg="#f0f8ff").pack()
entry_attendance = tk.Entry(root, width=25)
entry_attendance.pack()

tk.Label(root, text="Assignments Submitted:", bg="#f0f8ff").pack()
entry_assignments = tk.Entry(root, width=25)
entry_assignments.pack()

tk.Button(root, text="Predict Score", command=predict_score, bg="#4682B4", fg="white",
          font=("Arial", 11, "bold"), relief="ridge").pack(pady=15)

root.mainloop()
```
